```python
import random
class MyDataLoader2:
    def __init__(self, data1, data2):
        # data1과 data2 크기는 같다고 가정합시다
        self.size = len(data1) 
        self.data1 = data1 
        self.data2 = data2
       
 
    # __iter__에서는 나를 반환해주고 __next__에서는 for loop가
    # 돌떄마다 인쇄하고 싶은 값을 리턴합니다.
    def __iter__(self): 
        self.index = 0
        return self
 
    def __next__(self):
        
        if self.index >= self.size:
            raise StopIteration #일정기준 넘기면 종료
            
        n = self.data1[self.index] + self.data2[self.index]
        self.index += 1
        return n
    
        #data1과 data2를 더해서 리턴하는 iterator
        
    def shuffle(self):
        random.shuffle(self.data1)
        random.shuffle(self.data2)
        # data1과 data2를 shuffle
        
    def make_square(self):
        new_data1 = [i**2 for i in self.data1]
        new_data2 = [j**2 for j in self.data2]
        # data1과 data2의 값들을 다 제곱
        
```

```python

loader = MyDataLoader2([i for i in range(300)], [i for i in range(300)])
loader.make_square() ## data1과 data2를 제곱할 수 있는 method
loader.shuffle()
for i, x in enumerate(loader): # for loop 가 도니까 __iter__
    print("{} : {}".format(i,x)) # i번째 data1과 data2를 더한 값을 리턴
```
